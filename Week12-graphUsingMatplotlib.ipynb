```python
# Matplotlib를 사용하여 다양한 그래프 그리기
# import
from matplotlib import pyplot as plt
```

# matplotlib

```python
#plt.plot([x축 값], [y축 값], 옵션)
plt.plot([1,2,3,4])
plt.ylabel('y-axis values')
plt.title('graph') # graph title
plt.legend(['graph']) # legend
# annotation on graph xy : arrrow 위치 근데 안됨;;
# plt.annotate("info", xy=(1.5, 150), xytext=(2, 150), arrowprops{'color':'red'})
plt.show()
```

```python
plt.plot([1,2,3],[100,200,300],) # y축 100, 200, 300에 맞춰
plt.axis([1, 4, 100, 400]) # 각 축의 절대 크기 설정
plt.show()
```

```python
# 'y' - yellow 'r' - red 'ro' - marker
# ':ro' - dotted marker
plt.plot([1,2,3,4],[1,8,6,11],'ro')
plt.axis([0,6,0,12])
plt.show()
```

```python
plt.bar([1,2,3,4],[1,8,6,11]) # bargraph width=n (막대넓이)
plt.axis([0,5,0,12])
plt.show()
```

```python
plt.barh([1,2,3], [10,20,30]) # barhorizontal
plt.show()
```

```python
# 점 그래프 그리기
# x에 맞는 y축 점 그래프
plt.scatter([1,3,2,4,5],[10,5,20,30,1])
plt.show()
```

```python
# 원형 그래프
label = ['first', 'second', 'third', 'fourth']
plt.pie([10, 20, 30, 40], labels=label)
plt.show()
```

```python
# 한글 사용
# font_name=font_manager.FontProperties(fname="C:/windows/Fonts/malgun.ttf").get_name()
# rc('font', family=font_name)
# 전체 figure 크기를 10x10
plt.figure(figsize=(10,10))
# 1번째 plot 위치
plt.subplot(2,1,1)
plt.title('dot graph')
plt.plot([1,2,3,4,5], [1,8,6,11,12], 'ro')
plt.axis([0,6,0,13])
# 2번째 plot 위치
plt.subplot(2,1,2)
plt.title('bar graph')
plt.bar([1,2,3,4,5], [1,8,6,11,12], color='green')
plt.axis([0,6,0,13])
plt.show()
```

```python
from numpy import linspace, sqrt
# 1/sqrt(x**2+1) -1<=x<=5
x = linspace(-1,5,40)
y = 1.0/sqrt(x**2+1)
plt.scatter(x,y)

# x, y축 라벨
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
```

```python
# 두번째 방법
x = linspace(-1,5,40)
y = 1.0/sqrt(x**2+1)
# 땡땡이 그래프
plt.plot(x,y,'o')

# x, y축 라벨
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
```

```python
x = linspace(-1,5,50)
y1 = 1.0/sqrt(x**2+1)
y2 = 1.0/sqrt(3*x**2+1)
# 일반 그래프
plt.plot(x,y1)
# ------- 형식의 그래프
plt.plot(x,y2,'--')
plt.legend(['y1','y2'])
plt.show()
```

```python
# plot의 label을 이용해서도 그릴 수 있다.
x = linspace(-1,5,50)
y1 = 1.0/sqrt(x**2+1)
y2 = 1.0/sqrt(3*x**2+1)
# 일반 그래프
plt.plot(x,y1,label='y1')
# ------- 형식의 그래프
plt.plot(x,y2,'--',label='y2')
plt.show()
```

```python
from numpy import array
# 주어진 데이터를 이용하여 그래프 그리기
x = array([0,1,3,7,8])
y = array([1.1,2.5,-0.7,3.0,2.9])
plt.plot(x,y,'-or')
plt.xlabel('x')
plt.ylabel('y')
plt.title('grape')
plt.show()
```

```python
# 1번
import numpy as np
x = np.linspace(0,2.0*np.pi,50)
y = np.sin(x)+np.cos(x)
plt.plot(x,y,label='sin+cos')
plt.legend(['sin+cos'])
plt.title('sin+cos graph')
plt.ylabel('y-axis values')
plt.xlabel('x-axis values')
plt.show()
```

```python
# 2번
import numpy as np
x = np.linspace(0,2.0*np.pi,50)
y = np.sin(2*x)
plt.plot(x,y,label='sin2x')
plt.legend(['sin2x'])
plt.title('sin(2x) graph')
plt.ylabel('y-axis values')
plt.xlabel('x-axis values')
plt.show()
```

```python
# 3번
import numpy as np

x = linspace(0,3,50)
y = 1.0/np.sqrt(x**2+1)
plt.plot(x,y,label='sqrt')
plt.legend(['1.0/np.sqrt(x**2+1)'])
plt.title('1.0/np.sqrt(x**2+1) graph')
plt.ylabel('y-axis values')
plt.xlabel('x-axis values')
plt.show()
```

```python
# 4번
import numpy as np

x = np.linspace(-1,2.0*np.pi,50)
y1 = np.sin(x)
y2 = np.sin(2*x)
# 일반 그래프
plt.plot(x,y1)
# ------- 형식의 그래프
plt.plot(x,y2,'--')
plt.legend(['sin(x)','sin(2x)'])
plt.title('sin(x) and sin(2x) graph')
plt.ylabel('y-axis values')
plt.xlabel('x-axis values')
plt.show()
```

```python
# 5번
import numpy as np

x = np.linspace(-1,2.0*np.pi,50)
y1 = np.cos(x)
y2 = 2.0*np.sin(x)+1
# 일반 그래프
plt.plot(x,y1)
# ------- 형식의 그래프
plt.plot(x,y2,'--')
plt.legend(['cos(x)','2.0*np.sin(x)+1'])
plt.title('cos(x) and 2.0*np.sin(x)+1 graph')
plt.ylabel('y-axis values')
plt.xlabel('x-axis values')
plt.show()
```
